```python
def r2v(r):
    l = r.shape[0]
    return np.hstack([list(r[:, i].T) for i in range(l)])[0]

def v2r(v):
    n = int(np.sqrt(len(v)))
    return Qobj(np.hstack([c[..., None] for c in np.split(v,n)]))
```
